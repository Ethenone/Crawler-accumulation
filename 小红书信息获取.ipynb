```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.by import By
import urllib.request as req
from bs4 import BeautifulSoup as bsp
def openurl(url):
    res = req.Request(url)
    response = req.urlopen(res)
    content = response.read().decode("utf-8")
    return content

try:
    browser.quit()
except:
    print(1)
browser = webdriver.Edge(service=EdgeService(EdgeChromiumDriverManager().install()))

keyword = '繁花'
browser.get(f'https://www.xiaohongshu.com/search_result?keyword={keyword}')

# with open('cookies.txt', 'r') as file:  # 替换为你保存 Cookie 的文件路径
#     lines = file.readlines()
#     for line in lines[0].split(';'):
#         name, value = line.strip().split('=')
#         print(name,value)
#         browser.add_cookie({'name': name, 'value': value})

# browser.get(f'https://www.xiaohongshu.com/search_result?keyword={keyword}')




```

```python
sections = browser.find_elements(By.XPATH,'//section')

for section in sections:
    try:
        url = section.find_element(By.XPATH,'div/a[2]').get_attribute("href")
        print(url)
        soup = bsp(openurl(url),'lxml')
        print(soup.find('title').text)
        print(soup.find(attrs={"name": "keywords"}).get('content'))
        print(soup.find(attrs={"name": "description"}).get('content'))
    except:
        continue
```
