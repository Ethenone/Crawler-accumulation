```python
import pypinyin
def pinyin(word):
    s = []
    for i in pypinyin.pinyin(word, style=pypinyin.NORMAL):
        s.append(i[0])
    s = ' '.join(s)
    s = s.upper()
    return s
i = 0
retry = 0
while i < len(df):
    print(i)
    if pd.isnull(df.loc[i,'id']):
        i = i+1
        continue
    name = df.loc[i,'车站']
    py = pinyin(name)
    if py != df.loc[i,'拼音']:
        try:
            result = search(name)
            if result[2] != df.loc[i,'id']:
                print('wr at',i,'retry:',retry)
                result  = station(result[2],'i')
                info = result['info']
                df.loc[i,'拼音'] = info['sedName']
                df.loc[i,'路局'] = info['bureau']['name']
                df.loc[i,'地区'] = info['location']
                df.loc[i,'经度'] = info['x']
                df.loc[i,'纬度'] = info['y']
                df.loc[i,'id'] = info['id']
        except:
            if retry>5:
                i=i+1
                retry=0
                continue
            retry = retry + 1
            continue
    retry = 0
    i = i+1
            
```
