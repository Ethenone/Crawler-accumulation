```python
from urllib import parse
import urllib.request as req
import json
def openurl(url):
    res = req.Request(url)
    response = req.urlopen(res)
    content = response.read().decode("utf-8")
    return content
```

```python
# 查询接口
key = '上海'
key = parse.quote(key)
search_url = f'http://cnrail.geogv.org/api/v1/match_feature/{key}?locale=zhcn&query-override='
result = openurl(search_url)
result = json.loads(result)
print(result)
```

```python
# 车站查询接口
station_id = result["data"][0][0]
station_url = f'http://cnrail.geogv.org/api/v1/station/{station_id}?locale=zhcn&query-override=&requestGeom=true'
station_info = openurl(station_url)
station_info = json.loads(station_info)

#线路连接查询
link_url = f'http://cnrail.geogv.org/api/v1/station-link/{station_id}?locale=zhcn&query-override='
link = openurl(link_url)
link = json.loads(link)

#车次查询
trainlist_url = f'http://cnrail.geogv.org/api/v1/station_route/{station_id}?locale=zhcn&query-override='
trainlist = openurl(trainlist_url)
trainlist = json.loads(trainlist)

print(station_info,link,trainlist)
```

```python
# 线路查询接口
rail_id = result["data"][0][0]
rail_url = f'http://cnrail.geogv.org/api/v1/rail/{rail_id}?locale=zhcn'
rail_info = openurl(rail_url)
rail_info = json.loads(rail_info)
print(rail_info)
```

```python
# 车次信息查询
train_id = 'CN~1461'
train_url = f'http://cnrail.geogv.org/api/v1/route/{train_id}?locale=zhcn&query-override=&requestGeom=true'
train_info = openurl(train_url)
train_info = json.loads(train_info)
print(train_info)
```

```python
with open('车站名.txt','r',encoding='utf-8') as f:
    data = f.read()

stat_list = []
index = data.find('@',0)
while index != -1:
    index = index+5
    p = data.find('|',index)
    stat_list.append(data[index:p])
    index = data.find('@',index)

import pandas as pd
pd.DataFrame(stat_list).to_csv('车站.csv',encoding='gbk')
```
