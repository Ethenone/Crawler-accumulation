```python
import urllib.request as req
import json
res = req.Request(
    url = 'http://casad.cas.cn/ysxx2022/ysmd/sxwl/200906/t20090624_1792083.html')
response = req.urlopen(res)
content = response.read().decode("utf-8")
print(content)
```

```python
from bs4 import BeautifulSoup
html = content
soup = BeautifulSoup(html, 'lxml')
#print(soup)
print(soup.prettify())  # 使用prettify()格式化显示输出
```

```python
print(soup.select('.wztitle')[0].get_text())
```

```python
jlist = soup.find_all(name='p', align = 'justify')
justify = ""
for x in jlist:
    justify = justify + x.get_text()
print(justify)
```

```python
res = req.Request(
    url = 'http://casad.cas.cn/ysxx2022/ysmd/qtys/')
response = req.urlopen(res)
content = response.read().decode("utf-8")
print(content)
```

```python
html = content
soup = BeautifulSoup(html, 'lxml')
rlist = soup.select('.rmbs_a')
for x in rlist:
    hlist = x.find_all(name = 'a')
    for y in hlist:
        print(y.attrs["href"])
```
