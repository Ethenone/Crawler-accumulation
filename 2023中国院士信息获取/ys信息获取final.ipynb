```python
import urllib.request as req
from bs4 import BeautifulSoup as bsp
def openurl(url):
    res = req.Request(url)
    response = req.urlopen(res)
    content = response.read().decode("utf-8")
    return content
```

中科院

```python
url = 'http://casad.cas.cn/ysxx2022/ysmd/qtys/'
soup = bsp(openurl(url),'lxml')
dep = ["数学与物理学部","化学部","生命科学和医学学部","地学部","信息技术科学部","技术科学部"]
rlist = soup.select('.rmbs_a')
yslist = []
for r in rlist:
    hlist = r.find_all(name = 'a')
    ys = []
    for h in hlist:
        href = h.attrs["href"]
        soupx = bsp(openurl(href),'lxml')
        name = soupx.select('.wztitle')
        if len(name) == 0:
            continue
        name = name[0].get_text()
        justify = ""
        jlist = soupx.select('.acadTxt')[0].find_all(name = 'p')
        for j in jlist:
            justify = justify + j.get_text()
        ys.append([name,justify])
        print(name)
    yslist.append(ys)
print(yslist)
```

```python
import pandas as pd
for i in range(6):
    pd.DataFrame(yslist[i]).to_csv(f'{dep[i]}.csv',index=False,encoding='utf_8_sig')
```

工程院

```python
from tqdm import tqdm
url = 'https://www.cae.cn/cae/html/main/col48/column_48_1.html'
soup = bsp(openurl(url),'lxml')
hlist = soup.select('.name_list')
ys = []
for h in tqdm(hlist):
    href = 'https://www.cae.cn/' + h.a.attrs["href"]
    soupx = bsp(openurl(href),'lxml')
    name = soupx.select('.right_md_name')[0].get_text()
    justify = ""
    jlist = soupx.select('.intro')[0].find_all(name = 'p')
    for j in jlist:
        justify = justify + j.get_text()
    ys.append([name,justify])
pd.DataFrame(ys).to_csv('工程院院士.csv',index=False,encoding='utf_8_sig')
```
