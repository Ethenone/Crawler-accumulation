```python
import requests
import json
import pandas as pd
from tqdm import tqdm
```

```python
def getname(name):
    url = "https://dmfw.mca.gov.cn/9095/stname/listPub"
    data = {
        "page":1,
        "size":10,
        "stName":name,
        "year":0
    }
    res = requests.post(url=url,data=data)
    temp = json.loads(res.text)['records']
    if len(temp)==0:
        prov = 'None'
        city = 'None'
    else:
        prov = temp[0]['province_name']
        city = temp[0]['city_name']
    return prov,city
```

```python
data = pd.read_csv('出生地总表.csv')
pname = []
for d in tqdm(data['出生地']):
    prov,city = getname(d[-2:])
    pname.append([prov,city])
pd.DataFrame(pname).to_csv('出生地信息处理2.csv',index=False,encoding='utf_8_sig')
```

```python
def count(name):
    url = "https://dmfw.mca.gov.cn/9095/stname/count"
    data = {
        "page":1,
        "size":10,
        "stName":name,
        "year":0
    }
    res = requests.post(url=url,data=data)
    temp = json.loads(res.text)['def']
    if len(temp)==0:
        name = "None"
    else:
        name = temp[0]["name"]
    return name
```

```python
data = pd.read_csv('出生地总表.csv')
pname = []
for d in tqdm(data['出生地']):
    prov,city = getname(count(d[-2:]))
    pname.append([prov,city])
pd.DataFrame(pname).to_csv('出生地信息处理2.csv',index=False,encoding='utf_8_sig')
```
