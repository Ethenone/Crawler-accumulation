```python
import pandas as pd
data = pd.read_csv('工程院院士总表.csv')
print(data["justify"][0])
```

```python
from tqdm import tqdm
def pause(stri,index,a = "。",b = "，"):
    p1 = stri.find(a,index)
    p2 = stri.find(b,index)
    if p1 == -1:
        return p2
    elif p2 == -1:
        return p1
    else:
        return min(p1,p2)
def rpause(stri,index,a = "。",b = "，"):
    p1 = stri.rfind(a,0, index)
    p2 = stri.rfind(b,0, index)
    if p1 == -1:
        return p2
    elif p2 == -1:
        return p1
    else:
        return max(p1,p2)

info = []
i=0
for j in tqdm(data["justify"]):
    index = j.find("生于")
    if index == -1:
        index = j.find("人")
        if index == -1:
            birthplace = j
        else:
            p = rpause(j,index)
            birthplace = j[p+1:index]
    else:
        p = pause(j,index)
        birthplace = j[index+2:p]
    
    index = j.find("毕业于")
    if index == -1:
        university = j
    else:
        p = pause(j,index,a="大学",b="学院")
        university = j[index+3:p+2]
    
    index = j.find("当选")
    if index == -1:
        index = j.find("被选聘")
        if index == -1:
            index = j.find("选聘")
            if index == -1:
                year = j
            else:
                p = rpause(j,index)
                year = j[p+1:index]
        else:
            p = rpause(j,index)
            year = j[p+1:index]
    else:
        p = rpause(j,index)
        year = j[p+1:index]
    
    info.append([data["name"][i],birthplace,university,year])
    i = i + 1
pd.DataFrame(info).to_csv("info.csv",index=False,encoding='utf_8_sig')
```
